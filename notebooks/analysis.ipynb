## Change Financial
### Analysis of HMDA Data (2012 - 2014)

##### How should Change Financial enter into the home loans market?

Can Change Financial enter the home loan market by targeting the top markets with the most originated loans, to attract potential applicants that may be interested in refinancing their existing home loans? 

I hypothesize that this would be advantageous for Change Financial with the expectation that refinanced loans are relatively smaller in amount and thus are more affordable for the company. In addition the risks associated with these loans is relatively lower than that of purchase loans.  I believe that refinancing loans attract new customers in the top markets that have the potential to sign up for additional services.

```python
plot_asofyear(loans, 'State', top_ten=True)
```

Between 2012 and 2014 the number of loans originated in the Washington DC metroarea including Delaware, Virginia, West Virginia and Meryland has significantly declined. This is true for all the four states and District of Columbia although the rate of decline in the number of loans originated and the total value of the loans disbursed varies from state to state. Over the three years in our data, despite the decline, the largest number of loans originate in Virgina than any of the other states followed by Maryland. That is, the number of loans originating in Virgina have consistently stayed at 54 percent of the total loans through the years we have data for, while about 33 percent of the loans originate in Maryland. 

We can see from these numbers that close to 90 percent of the home loans market is concentrated in the two states, Maryland and Virginia. Remember that since Change Financial is only planning to enter the home loans market in the DC metro area, we are exploring the data to see if it is best to strictly focus on the top markets at this point. We can further narrow down and see whether there is a concentration of loans in counties or metropolitan statistical areas within the states.

```python
plot_asofyear(loans, 'County_Name', top_ten=True)
```

Looking at the graph above, which shows the breakdown of the volume, total loan amount and median loan amount broken down by county, we can see that most of the loans originate in a few counties in Virginia and Maryland. For instance, from the top 20 counties with the most number of originating loans, five counties in Maryland and Virginia make up about 42 percent. This clearly shows us that the housing market in the Washington DC metro area is concentrated in a few counties and states.

```python
plot_asofyear(loans, 'MSA_MD_Description', top_ten=True)
```

The concentration of the home loans market is further evidened by the graph above which shows a breakdown of loans volume, total loan amount and median loan amount broken down by metropolitan statistical area.

```python
plot_loan_purpose(loans,'State', xlab='State', title='Loans Volume for Each State Broken Down by Loan Purpose (2012 - 2014)')
```

```python
plot_loan_purpose(loans,'State', year=2014, xlab='State', title='Loans Volume for Each State Broken Down by Loan Purpose (2014)')
```

The graphs above show a breakdown of volume of loans broken down by applicants' purpose for the loan (i.e. refinancing an existing loan or for the purchase of buying a new home) for all years and for the year 2014 respectively. The data presented in the graphs are important because we are hypothesizing that loans for refinancing existing loans are relatively low in risk and smaller in amount than new loans. Although we cannot generalize with absolute certainity the level of risk in each type of loan, in terms of purpose, by looking at the graph above, we can however make an educated guess that refinance loans are less risky from the fact that in all of the five states in our data, a significantly large share of the loans are for refinancing than purchase. That is, lending institutions who are already in the market seem to approve refinancing loans more than new home purchase loans. This is especially true for the two states, Virginia and Maryland, with the largest share in the home loans market. 

```python
plot_asofyear(loans, 'Loan_Purpose_Description', top_ten=True)
```

We can also observe below that the median loan amount for refinance loans are slightly smaller in amount than purchase loans. 

```python
plot_loan_purpose(loans, 'Respondent_Name_TS', xlab='Institution', title='Top 20 Institutions - Loans Volume Break Down by Purpose')
```

We can see further evidence above that existing lending institutions with the highest market share prefer to extend refinance loans than new loans for a purchase of a home supporting our hypothesis to target refinance loans in the largest markets. 

We saw from our analysis that existing lenders with the largest market share operate mostly in areas with the highest volume of loans and they originate refinance loans in significantly high rates than new loans for purchase of a home. We also saw that the median refinance loan is merely slightly smaller in amount than purchase loans, which does not support our hypothesis that much. Yet, it it apparent that if we believe it is a good idea to emulate the lending patterns of institutions with the largest market share, then it means Change Financial would lend greatly more in refinance loans in the largest markets.
